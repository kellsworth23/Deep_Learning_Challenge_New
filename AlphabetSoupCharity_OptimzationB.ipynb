## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
url = 'https://raw.githubusercontent.com/kellsworth23/Deep_Learning_Challenge_New/main/Resources/charity_data.csv'
charity_df = pd.read_csv(url)
charity_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN'.
charity_df = charity_df.drop(columns = ['EIN'])
```

```python
# Determine the number of unique values in each column.
for x in charity_df.columns:
    print(x, len(charity_df[x].unique()))
```

```python
# Look at NAME counts for binning
counts_name = charity_df['NAME'].value_counts()
counts_name
```

```python
# Choose a cutoff value and create a list of names_replace to be replaced
# use the variable name 'names_to_replace'
names_to_replace = list(counts_name[counts_name < 50].index)
names_to_replace
```

```python
# Replace in dataframe
for name in names_to_replace:
    charity_df['NAME'] = charity_df['NAME'].replace(name,"Other")

# Check to make sure binning was successful
charity_df['NAME'].value_counts()
```

```python
# Look at APPLICATION_TYPE value counts for binning
counts = charity_df['APPLICATION_TYPE'].value_counts()
counts
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
application_types_to_replace = list(counts[counts < 700].index)
application_types_to_replace

# Replace in dataframe
for app in application_types_to_replace:
    charity_df['APPLICATION_TYPE'] = charity_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
charity_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
counts_for_binning = charity_df['CLASSIFICATION'].value_counts()
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
classification_counts = counts_for_binning[counts_for_binning > 1]
classification_counts
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
classifications_to_replace = list (counts_for_binning[counts_for_binning < 100].index)
classifications_to_replace

# Replace in dataframe
for cls in classifications_to_replace:
    charity_df['CLASSIFICATION'] = charity_df['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
charity_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
charity_df = pd.get_dummies(charity_df,dtype=float)
charity_df.head()
```

```python
# Split our preprocessed data into our features and target arrays
y = charity_df['IS_SUCCESSFUL'].values
y

X = charity_df.drop('IS_SUCCESSFUL', axis=1).values
X
# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state = 42)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
number_input_features = len( X_train_scaled[0])
hidden_nodes_layer1=8
hidden_nodes_layer2=15
nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation='relu'))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation='relu'))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics=['accuracy'])
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled,y_train,validation_split=0.15, epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
